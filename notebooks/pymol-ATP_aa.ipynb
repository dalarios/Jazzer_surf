```python
import pymol2
import csv
import os
import pandas as pd

def analyze_atp_binding(protein_file, ligand_name="ATP", distance_threshold=5.0):
    with pymol2.PyMOL() as pymol:
        pymol.cmd.load(protein_file)
        pymol.cmd.select("ligand", f"resn {ligand_name}")
        pymol.cmd.select("binding_residues", f"byres (ligand around {distance_threshold})")
        
        atp_atoms = pymol.cmd.get_model("ligand").atom
        binding_residues = pymol.cmd.get_model("binding_residues").atom

        interactions = []

        for atp_atom in atp_atoms:
            atp_selection = f"chain {atp_atom.chain} and resi {atp_atom.resi} and name {atp_atom.name}"
            for residue_atom in binding_residues:
                residue_selection = f"chain {residue_atom.chain} and resi {residue_atom.resi} and name {residue_atom.name}"
                
                try:
                    distance = pymol.cmd.get_distance(atp_selection, residue_selection)
                    if distance <= distance_threshold:
                        interactions.append((atp_atom.resn, atp_atom.resi, atp_atom.name,
                                             residue_atom.resn, residue_atom.resi, residue_atom.name, distance))
                except pymol.CmdException as e:
                    print(f"Error measuring distance between {atp_selection} and {residue_selection}: {e}")

        return interactions

def process_files_in_folder(folder_path, ligand_name="ATP", distance_threshold=5.0, output_csv=None):
    all_interactions = []

    for filename in os.listdir(folder_path):
        if filename.endswith(".pdb") or filename.endswith(".cif"):
            protein_file = os.path.join(folder_path, filename)
            print(f"Processing file: {protein_file}")
            interactions = analyze_atp_binding(protein_file, ligand_name, distance_threshold)
            for interaction in interactions:
                all_interactions.append([filename] + list(interaction))

    if output_csv is None:
        output_csv = os.path.join(folder_path, "interactions_combined.csv")

    # Write all interactions to a single CSV file
    os.makedirs(os.path.dirname(output_csv), exist_ok=True)
    with open(output_csv, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Protein File", "ATP Residue Name", "ATP Residue ID", "ATP Atom Name", "Binding Residue Name", "Binding Residue ID", "Binding Atom Name", "Distance (Å)"])
        for interaction in all_interactions:
            writer.writerow(interaction)

    print(f"All interactions have been saved to {output_csv}")

    # Create a DataFrame from the interactions
    df = pd.DataFrame(all_interactions, columns=["Protein File", "ATP Residue Name", "ATP Residue ID", "ATP Atom Name", "Binding Residue Name", "Binding Residue ID", "Binding Atom Name", "Distance (Å)"])

    # Save the DataFrame to a CSV file
    dataframe_output_path = os.path.join(folder_path, "interactions_combined_dataframe.csv")
    df.to_csv(dataframe_output_path, index=False)

    print(f"DataFrame has been saved to {dataframe_output_path}")

    return df

# Example usage
folder_path = "/Users/dalarios/git/Jazzer_surf/3d_predictions"
df = process_files_in_folder(folder_path)
print(df.head())
```
